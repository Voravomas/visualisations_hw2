```python
import pandas as pd
import altair as alt
import datetime as dt
alt.data_transformers.disable_max_rows()
```

```python
def load_data():
    return pd.read_excel('2014-2020.xlsx')

df = load_data()
# df.head()
df[:30]
```

1. Як змінювалась структура генерації електроенергії за роками?

```python
# 1. Preprocess data
# get sum of energy generation per type per year

new_df = {"year": [],
          "type": [],
          "mvt": []
         }

all_types = ["AES", "TEC", "VDE", "TES", "GES", "GAES_GEN"]

for year in range(2014, 2021):
    year_df = df[df['Час/Дата'].str.contains(str(year))]
    for exact_type in all_types:
        new_df["year"].append(str(year))
        new_df["type"].append(exact_type)
        sum_of_electricity = sum(list(year_df[exact_type].fillna(0)))
        new_df["mvt"].append(sum_of_electricity)

new_df = pd.DataFrame.from_dict(new_df)
print(new_df)
```

```python
# 2. visualise

alt.Chart(new_df).mark_bar().encode(
    x=alt.X('type:O', title=None),
    y='mvt:Q',
    color='type:N',
    column='year:N',
).properties(
    title='year',
).configure_title(
    orient='bottom',
    anchor='middle',
)
```

Перший який прийшов в голову

**Які ще альтернативи ви розглядали?**
Можна було ще зробити 6 окремих графіків (типи) розбиті по роках 

**Переваги цього варіанту?**
К-сть мегаватів показується один раз, так як питання саме у цьому

**Недоліки цього варіанту?**
- Треба препроцесити дані
- Широкий
- Зміну малих одиниць важко розлічити


2. Як залежить споживання електроенергії від дня року та години доби?

```python
# 1. Preprocess data

# hour, day, sum(mvt)
# Час/Дата Час/Дата, average(consumption)


# "24-31.12.2020"

def extract_year(x):
    return int(x[-4:])

def extract_month(x):
    return int(x.split(".")[1])

def extract_day(x):
    return int(x.split(".")[0].split("-")[1])

def extract_hour(x):
    return int(x.split("-")[0])

def get_weekday(x):
    d_m_y = [int(elm) for elm in x.split("-")[1].split(".")]
    return dt.datetime(d_m_y[2], d_m_y[1], d_m_y[0]).weekday()




df["year"], df["month"] = df["Час/Дата"], df["Час/Дата"]
df["day"], df["hour"] = df["Час/Дата"], df["Час/Дата"]
df["weekday"] = df["Час/Дата"]

df["year"] = df["year"].apply(extract_year)
df["month"] = df["month"].apply(extract_month)
df["day"] = df["day"].apply(extract_day)
df["hour"] = df["hour"].apply(extract_hour)
df["weekday"] = df["weekday"].apply(get_weekday)

df.head()
```

```python
def mask(df, key, value):
    return df[df[key] == value]

new_df = {"day": [], "hour": [], "avg_mvt": []}
filtered_df = df.filter(["year", "month", "day", "hour", "CONSUMPTION"])

day_of_year = 1
for month in range(1, 13):
    for day in range(1, 32):
        for hour in range(1, 25):
            values = mask(mask(mask(filtered_df, "month", month), "day", day), "hour", hour)
            if values.empty:
                break
            new_df["day"].append(day_of_year)
            new_df["hour"].append(hour)
            avg_value = round(sum(values["CONSUMPTION"])/len(values))
            new_df["avg_mvt"].append(avg_value)
        if not values.empty:
            day_of_year += 1
            

new_df = pd.DataFrame.from_dict(new_df)
new_df.head()

```

```python
# 2. Visualise

alt.Chart(new_df).transform_filter(alt.FieldRangePredicate(field='hour', range=[1, 24])).mark_rect().encode(
    x = alt.X('hour:O'),
    y = alt.Y('day:N', sort = alt.Sort(order = 'descending')),
    color = alt.Color('avg_mvt:Q', scale = alt.Scale(scheme = 'blueorange'))
)
```

**Чому ви обрали саме цей спосіб презентації даних?**
Як на мене показати цю візуалізацію було досить складно, а даний тип візуалізації вже був в прикладах

**Які ще альтернативи ви розглядали?**
Єдине що мені прийшло на думку

**Переваги цього варіанту?**
- показує залежність дня і години

**Недоліки цього варіанту?**
- довгий

4. Як змінюється споживання електроенергії впродовж доби у розрізі місяців року та пір року?

```python
# 3. Preprocess season

season_dict = {"winter": [1, 2, 12],
               "spring": [3, 4, 5],
               "summer": [6, 7, 8],
               "autumn": [9, 10, 11]}



new_df = {"season": [],
          "hour": [],
          "CONSUMPTION_visual": [],
          "CONSUMPTION_real": [],
         }

MINUMUM_CONSUMPTION_AVG = 393963  # precalculated

for hour in range(1, 25):
        hour_values = mask(df, "hour", hour)
        for season_name, season_month in season_dict.items():
            season_consumption = 0
            for month_number in season_month:
                month_values = mask(hour_values, "month", month_number)
                cons_value = round(sum(month_values["CONSUMPTION"])/len(values))
                season_consumption += cons_value
            new_df["season"].append(season_name)
            new_df["hour"].append(hour)
            new_df["CONSUMPTION_visual"].append(round(season_consumption / 3) - MINUMUM_CONSUMPTION_AVG)
            new_df["CONSUMPTION_real"].append(round(season_consumption / 3))

            
new_df = pd.DataFrame.from_dict(new_df)
new_df.head()
```

```python
# 4. Visualize season

alt.Chart(new_df).mark_point().encode(
    x = alt.X('hour:Q'),
    y = alt.Y('season:N', sort=["winter", "spring", "summer", "autumn"]),
    size = alt.Size('CONSUMPTION_visual', scale = alt.Scale(range = [100, 1000])),
    tooltip = alt.Tooltip('CONSUMPTION_real')
).properties(width = 950, height = 500)
```

5. Як змінюється споживання електроенергії впродовж тижня?

```python
# 1. Preprocess

new_df = {"weekday": [0, 1, 2, 3, 4, 5, 6], "avg_mvt": []}

for wd in new_df["weekday"]:
    needed_day_data = mask(df, "weekday", wd).fillna(0)
    new_df["avg_mvt"].append(round(needed_day_data["CONSUMPTION"].sum() / len(needed_day_data)))
    

new_df = pd.DataFrame.from_dict(new_df)
print(new_df)
```

```python
# 2. Visualize

alt.Chart(new_df).mark_line().encode(
    x="weekday:Q",
    y=alt.Y(field = 'avg_mvt', 
              type = 'quantitative',
              scale = alt.Scale(
                  zero = False
              ))
).properties(width = 800, height = 500)
```
